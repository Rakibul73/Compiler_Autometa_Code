```python
import re

def extract_comments(code):
    # Regular expression for single-line comments (//)
    single_line_pattern = r'\/\/.*'

    # Regular expression for multi-line comments (/* ... */)
    multi_line_pattern = r'\/\*.*?\*\/'

    # Find single-line comments
    single_line_comments = re.findall(single_line_pattern, code)

    # Find multi-line comments
    multi_line_comments = re.findall(multi_line_pattern, code, re.DOTALL)

    return single_line_comments + multi_line_comments

# Example C/C++ code
c_code = """
// This is a single-line comment
int main() {
    /*
    This is a
    multi-line
    comment */
    return 0;
}
"""


import re

def is_valid_identifier(identifier):
    # Check if the identifier starts with a letter or underscore
    if not re.match(r'^[a-zA-Z_]', identifier):
        return False

    # Check if the remaining characters are letters, digits, or underscores
    if not re.match(r'^[a-zA-Z0-9_]*$', identifier):
        return False

    # Check if the identifier is a reserved word for python 
    reserved_words = set(['if', 'else', 'while', 'for', 'int', 'float', 'return', 'def', 'class', 'import', 'from'])
    if identifier in reserved_words:
        return False
    
    # Check if the identifier is a reserved word for c++
    reserved_words_c_and_cpp = set([
        'auto', 'break', 'case', 'char', 'const', 'continue', 'default', 'do', 'double', 'else', 
        'enum', 'extern', 'float', 'for', 'goto', 'if', 'int', 'long', 'register', 'return', 
        'short', 'signed', 'sizeof', 'static', 'struct', 'switch', 'typedef', 'union', 'unsigned', 
        'void', 'volatile', 'while',
        
        # Additional keywords in C++
        'bool', 'class', 'const_cast', 'delete', 'dynamic_cast', 'explicit', 'export', 'false', 
        'friend', 'inline', 'mutable', 'namespace', 'new', 'operator', 'private', 'protected', 
        'public', 'reinterpret_cast', 'static_cast', 'template', 'this', 'throw', 'true', 'try', 
        'typeid', 'typename', 'using', 'virtual', 'wchar_t'
    ])
    if identifier in reserved_words_c_and_cpp:
        return False


    return True


def recognize_whitespace_in_python(file_path):
    # Initialize counts
    blank_count = 0
    # whitespace_count = 0
    tab_count = 0
    newline_count = 0

    try:
        # Read the Python code from the file
        with open(file_path, 'r', encoding='utf-8') as file:
            python_code = file.read()

        # Iterate through each character in the Python code
        for char in python_code:
            if char == ' ':
                blank_count += 1
            # elif char.isspace():
            #     whitespace_count += 1
            elif char == '\t':
                tab_count += 1
            elif char == '\n':
                newline_count += 1

    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

    return blank_count, tab_count, newline_count

import re

def lexical_analyzer(input_string):
    # Arithmetic Operators
    arithmetic_operators = ['+', '-', '*', '/', '//', '%', '**']
    # Assignment Operators
    assignment_operators = ['=', '+=', '-=', '*=', '/=', '//=', '%=', '**=', '&=', '|=', '^=', '<<=', '>>=']
    # Comparison Operators
    comparison_operators = ['==', '!=', '<', '>', '<=', '>=']
    # Logical Operators
    logical_operators = ['and', 'or', 'not']
    # Bitwise Operators
    bitwise_operators = ['&', '|', '^', '~', '<<', '>>']
    # Membership Operators
    membership_operators = ['in', 'not in']
    # Identity Operators
    identity_operators = ['is', 'is not']

    operators = arithmetic_operators + assignment_operators + comparison_operators + \
                logical_operators + bitwise_operators + membership_operators + \
                identity_operators


    # Use regular expression to tokenize the input string
    tokens = re.findall(r'\b\w+\b|[\+\-\*/=<>!]+', input_string)
    print(tokens)

    # Validate and print the operators
    for token in tokens:
        if token in operators:
            print(f"Valid operator: {token}")
        else:
            print(f"Invalid operator: {token}")
    


if __name__ == "__main__":

    type = input('Commemnts-> 1 Constants-> 2 Identifier-> 3 Operator-> 4 White space-> 5')
 
    if type == '1':
        # Extract comments from the code
        comments = extract_comments(c_code)

        # Print the extracted comments
        for comment in comments:
            print(comment.strip())

    if type == '2':
        user_input = input('Insert to check Constants : ')
        pas = 0
        for ch in user_input:
            if ch >= '0' and ch <= '9':
                continue
            pas = 1
        if pas:
            print("Not a constant");
        else:
            print("A constant");
            

    if type == '3':
        #identifier
        user_input = input('Insert to check identifier : ')

        if is_valid_identifier(user_input):
            print(f"The identifier '{user_input}' is valid.")
        else:
            print(f"The identifier '{user_input}' is not valid.")

    if type == '4':
        input_expr = input("Enter an expression with operators: ")
        print(input_expr)

        # Call the lexical analyzer function
        lexical_analyzer(input_expr)

    if type == '5':
        # Hardcoded file path for testing
        hardcoded_file_path = "example.c"

        # Call the function to recognize whitespace in the Python code
        blank, tab, newline = recognize_whitespace_in_python(hardcoded_file_path)

        # Display the results
        print(f"Occurrences of White spaces: {blank}")
        # print(f"Occurrences of white spaces, tabs, and newlines: {whitespace}")
        print(f"Occurrences of tabs: {tab}")
        print(f"Occurrences of newlines: {newline}")

```
