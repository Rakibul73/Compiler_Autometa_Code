
Implementing bottom-up parsing involves using a parsing table and a stack to reduce the input string to the start symbol. One common bottom-up parsing algorithm is the LR parser

```python
gram = {
	"S":["CC"],
	"C":["aC","d"]
}
start = "S"
terms = ["a","d","$"]

non_terms = []
for i in gram:
	non_terms.append(i)
gram["S'"]= [start]


new_row = {}
for i in terms+non_terms:
	new_row[i]=""


non_terms += ["S'"]
# each row in state table will be dictionary {nonterms ,term,$}
stateTable = []
# I = [(terminal, closure)]
# I = [("S","A.A")]

def Closure(term, I):
	if term in non_terms:
		for i in gram[term]:
			I+=[(term,"."+i)]
	I = list(set(I))
	for i in I:
		# print("." != i[1][-1],i[1][i[1].index(".")+1])
		if "." != i[1][-1] and i[1][i[1].index(".")+1] in non_terms and i[1][i[1].index(".")+1] != term:
			I += Closure(i[1][i[1].index(".")+1], [])
	return I

Is = []
Is+=set(Closure("S'", []))


countI = 0
omegaList = [set(Is)]
while countI<len(omegaList):
	newrow = dict(new_row)
	vars_in_I = []
	Is = omegaList[countI]
	countI+=1
	for i in Is:
		if i[1][-1]!=".":
			indx = i[1].index(".")
			vars_in_I+=[i[1][indx+1]]
	vars_in_I = list(set(vars_in_I))
	# print(vars_in_I)
	for i in vars_in_I:
		In = []
		for j in Is:
			if "."+i in j[1]:
				rep = j[1].replace("."+i,i+".")
				In+=[(j[0],rep)]
		if (In[0][1][-1]!="."):
			temp = set(Closure(i,In))
			if temp not in omegaList:
				omegaList.append(temp)
			if i in non_terms:
				newrow[i] = str(omegaList.index(temp))
			else:
				newrow[i] = "s"+str(omegaList.index(temp))
			print(f'Goto(I{countI-1},{i}):{temp} That is I{omegaList.index(temp)}')
		else:
			temp = set(In)
			if temp not in omegaList:
				omegaList.append(temp)
			if i in non_terms:
				newrow[i] = str(omegaList.index(temp))
			else:
				newrow[i] = "s"+str(omegaList.index(temp))
			print(f'Goto(I{countI-1},{i}):{temp} That is I{omegaList.index(temp)}')

	stateTable.append(newrow)
print("\n\nList of I's\n")
for i in omegaList:
	print(f'I{omegaList.index(i)}: {i}')


#populate replace elements in state Table
I0 = []
for i in list(omegaList[0]):
	I0 += [i[1].replace(".","")]
print(I0)

for i in omegaList:
	for j in i:
		if "." in j[1][-1]:
			if j[1][-2]=="S":
				stateTable[omegaList.index(i)]["$"] = "Accept"
				break
			for k in terms:
				stateTable[omegaList.index(i)][k] = "r"+str(I0.index(j[1].replace(".","")))
print("\nStateTable")

print(f'{" ": <9}',end="")
for i in new_row:
	print(f'|{i: <11}',end="")

print(f'\n{"-":-<66}')
for i in stateTable:
	print(f'{"I("+str(stateTable.index(i))+")": <9}',end="")
	for j in i:
		print(f'|{i[j]: <10}',end=" ")
	print()
```

```python
"""
This code uses a simple stack-based approach to implement
a bottom-up parser for a given grammar. The grammar is
defined as a dictionary, where the keys are the non-terminal
symbols and the values are lists of their possible expansions.
"""

# Define the grammar in the form of a dictionary
# The keys are the non-terminal symbols, and the values are lists of their possible expansions
grammar = {
    'S': [['NP', 'VP']],
    'NP': [['Det', 'Noun'], ['PN']],
    'VP': [['Verb', 'NP'], ['Verb']],
    'Det': [['the'], ['a']],
    'Noun': [['cat'], ['dog'], ['man'], ['woman']],
    'PN': [['John'], ['Mary']],
    'Verb': [['chases'], ['likes'], ['hates']]
}

# Define the input sentence as a list of tokens
sentence = ['John', 'hates', 'the', 'woman']

# Initialize the stack with the end-of-sentence marker and the initial symbol
stack = ['$', 'S']

# Initialize the input pointer to the first token
input_pointer = 0

# Loop until the stack is empty
while stack:
    # Get the top symbol on the stack
    top_symbol = stack.pop()

    # If the top symbol is a terminal, check if it matches the current input token
    if top_symbol in ['the', 'a', 'cat', 'dog', 'man', 'woman', 'John', 'Mary', 'chases', 'likes', 'hates', '$']:
        if top_symbol == sentence[input_pointer]:
            input_pointer += 1
        else:
            # If the top symbol does not match the current input token, the input is not valid
            print("Error: invalid input")
            break

    # If the top symbol is a non-terminal, expand it using the grammar
    elif top_symbol in grammar:
        expansion = grammar[top_symbol].pop(0)
        stack.extend(reversed(expansion))

# If the input pointer is at the end of the input and the stack is empty, the input is valid
if input_pointer == len(sentence) and not stack:
    print("Input is valid")
```
